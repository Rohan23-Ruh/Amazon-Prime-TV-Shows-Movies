## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Remove duplicate rows from the "ti" DataFrame
# Exclude 'genres' and 'production_countries' from subset for drop_duplicates if they are lists
# This handles cases where these columns might already be lists from previous executions.
hashable_cols = [col for col in ti.columns if col not in ['genres', 'production_countries']]
ti.drop_duplicates(subset=hashable_cols, inplace=True)

# Convert 'genres' and 'production_countries' columns from string representations of lists to actual Python lists
# Only apply ast.literal_eval if the column is of string type (i.e., not already a list)
for col in ['genres', 'production_countries']:
    if ti[col].apply(type).eq(str).any(): # Check if any value is a string
        ti[col] = ti[col].apply(lambda x: ast.literal_eval(x))

# Handle missing values in "ti"
ti['age_certification'].fillna('Unknown', inplace=True)
ti['description'].fillna('No Description', inplace=True)
ti['seasons'].fillna(0, inplace=True) # Assuming 0 seasons for movies or unknown shows

# Merge the "ti" (titles) and "df" (credits) DataFrames into a new 'merged_df'
merged_df = pd.merge(ti, df, on='id', how='left')

# Further Data Wrangling for merged_df
# Handling missing 'imdb_id'
merged_df['imdb_id'].fillna('Unknown_ID', inplace=True)

# Handling missing numerical score and popularity columns with median imputation
numerical_cols_to_impute = ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
for col in numerical_cols_to_impute:
    if merged_df[col].isnull().any():
        median_val = merged_df[col].median()
        merged_df[col].fillna(median_val, inplace=True)

# Handling missing values in 'person_id', 'name', 'character', 'role'
# Fill 'person_id' with a distinct numerical placeholder (-1)
merged_df['person_id'].fillna(-1, inplace=True)
merged_df['name'].fillna('Unknown_Name', inplace=True)
merged_df['character'].fillna('Unknown_Character', inplace=True)
merged_df['role'].fillna('Unknown_Role', inplace=True)

# Convert 'seasons' to integer type after filling NaNs with 0
merged_df['seasons'] = merged_df['seasons'].astype(int)

# Display a sample of the cleaned merged_df and its information to verify
print("Info of merged_df after all data wrangling steps:")
merged_df.info()
print("\nFirst 5 rows of the wrangled merged_df:")
display(merged_df.head())
````

### What all manipulations have you done and insights you found?

I have performed the following data manipulations:

1.  **Removed Duplicate Rows**: Identified and removed 3 duplicate rows from the **ti** DataFrame. This ensures that each title entry is unique, preventing skewed analyses.
2.  **Converted String-Represented Lists**: The **genres** and **production_countries** columns, which were strings representing lists, have been converted into actual Python lists using **ast.literal_eval**. This allows for easier manipulation and analysis of individual genres and countries.
3.  **Handled Missing Values in ti DataFrame**:
    *   **age_certification**: Filled missing values with 'Unknown'.
    *  ** description**: Filled missing values with 'No Description'.
    *   **seasons**: Filled missing values with 0, assuming movies or unknown for missing season counts.
4.  **Merged DataFrames**: The **ti** (titles) DataFrame and **df** (credits) DataFrame were merged using a **left** merge on the common **id **column. The resulting **merged_df** now contains comprehensive information about each title, including its cast and crew.
5.  **Further Handled Missing Values in merged_df**:
    *   **imdb_id**: Filled missing values with 'Unknown_ID' to provide a consistent identifier.
    *  ** imdb_score, imdb_votes, tmdb_popularity, `tmdb_score**: Filled missing numerical values with their respective *medians*. This is a robust imputation strategy that helps retain data for analysis while being less sensitive to outliers than mean imputation.
    *   **person_id:** Filled missing person_id values with -1 as a numerical placeholder for unknown credit entries.
    *   **name, character, role**: Filled missing values with 'Unknown_Name', 'Unknown_Character', and 'Unknown_Role' respectively, to indicate an absence of credit information for those entries.
6.  **Type Conversion**: The seasons column was converted to an integer type after all its missing values were filled with 0, making its data type consistent with season counts.

These manipulations significantly improve the data quality, consistency, and completeness of the merged_df, making it thoroughly prepared for comprehensive exploratory data analysis and visualization.

#### Chart - 10

````python
# Chart - 10 Number of Titles Released Per Year by Top 5 Production Countries

# Flatten the list of production countries and count their occurrences
all_countries = [country for sublist in merged_df['production_countries'] for country in sublist]
country_counts = pd.Series(all_countries).value_counts()

# Select the top 5 countries for visualization
top_5_countries = country_counts.head(5).index.tolist()

# Filter the merged_df for content from these top countries
top_countries_df = merged_df[merged_df['production_countries'].apply(lambda x: any(country in x for country in top_5_countries))].copy()

# Create a new column 'country' which will hold the main country for plotting (handling multi-country productions simplistically for this chart)
# For simplicity, if a title has multiple countries, we'll assign it to the first country in its list that is among the top 5.
def get_main_country(country_list, top_countries):
    for country in country_list:
        if country in top_countries:
            return country
    return None

# Create a new column containing the selected main production country
top_countries_df['main_country'] = top_countries_df['production_countries'].apply(lambda x: get_main_country(x, top_5_countries))

# Drop rows where main_country is None (titles not primarily from top 5)
top_countries_df.dropna(subset=['main_country'], inplace=True)

# Group by release_year and main_country to count titles per year per country
country_year_counts = top_countries_df.groupby(['release_year', 'main_country']).size().unstack(fill_value=0).reset_index()
# Create a figure with specified size
plt.figure(figsize=(15, 8))
# Plot release trends for each top production country
for country in top_5_countries:
    if country in country_year_counts.columns:
        sns.lineplot(data=country_year_counts, x='release_year', y=country, label=country, marker='o')

# Add title to the chart
plt.title('Number of Titles Released Per Year by Top 5 Production Countries')

# Label x-axis/y-axis
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
# Display legend
plt.legend(title='Production Country')
# Add grid lines for readability
plt.grid(axis='both', linestyle='--', alpha=0.7)
# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')
# Adjust layout to avoid overlapping
plt.tight_layout()
# Display the chart
plt.show()
````

##### 1. Why did you pick the specific chart?

A line plot is selected to effectively visualize trends over time, showing how the number of titles produced by the top countries has changed across different release years. This chart is ideal for comparing the production volume trajectories of multiple categories (countries) over a continuous variable (time), highlighting periods of growth, decline, or stability for each. It helps in understanding the historical contribution of different nations to Amazon Prime Video's content library.

##### 2. What is/are the insight(s) found from the chart?

The line chart clearly illustrates the dominance of certain countries in content production. The **United States (US)** consistently leads by a significant margin in terms of the number of titles produced each year, showing a strong upward trend, particularly from the early 2000s onwards, with a peak around 2017-2019. Other countries like **India (IN)** and the **United Kingdom (GB)** also show notable contributions, with their production numbers increasing over time, though far less than the US.
